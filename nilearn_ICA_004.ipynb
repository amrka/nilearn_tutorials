## <font color=green>8.4.1. Group analysis of resting-state fMRI with ICA: CanICA

```python
from nilearn import datasets, plotting, image
```

```python
adhd_dataset = datasets.fetch_adhd(n_subjects=30)
func_filenames = adhd_dataset.func

print('First functional nifti image (4D) is at: %s '%func_filenames[0])
```

```python
from nilearn.decomposition import CanICA
```

```python
canica = CanICA(n_components=20, smoothing_fwhm=6.,
               memory='nilearn_cache', memory_level=2, threshold=3., verbose=10, random_state=0)

canica.fit(func_filenames)
```

```python
components_img = canica.components_img_

components_img.to_filename('canica_resting_state.nii.gz')
```

```python

plotting.plot_glass_brain(image.index_img(components_img, 0))
```

```python
plotting.plot_prob_atlas(components_img, title='All ICA components')
```

```python
for i, cur_img in enumerate(image.iter_img(components_img)):

    plotting.plot_stat_map(cur_img, display_mode="z", title='IC %d'%i, cut_coords=1, colorbar=False)
```

## 3.3.2. Beyond ICA : Dictionary learning

```python
from nilearn import datasets
adhd_dataset = datasets.fetch_adhd(n_subjects=30)
func_filenames = adhd_dataset.func
```

```python
from nilearn.decomposition import DictLearning, CanICA

n_components = 40
```

```python
dict_learning = DictLearning(n_components=n_components,
                            memory='nilearn_cache', memory_level=2,
                            verbose=1, random_state=0, 
                            n_epochs=1, mask_strategy='template')
```

```python
canica = CanICA(n_components=n_components,
               memory='nilearn_cache', memory_level=2,
               threshold=3.,
               n_init=1, verbose=1,
               mask_strategy='template')
```

```python
estimators = [dict_learning, canica]
names = {dict_learning: 'DictLearning', canica: 'CanICA'}
components_imgs = []
```

```python
for estimator in estimators:
    print('[Example] learning maps using %s model' % names[estimator])
    estimator.fit(func_filenames)
    print('[Example] Saving results')
    components_img = estimator.components_img_
    components_img.to_filename('{0}_resting_stat.nii.gz'.format(names[estimator]))
    components_imgs.append(components_img)
```

```python
indices = {dict_learning: 25, canica: 33}
cut_components = image.index_img(components_img[0], indices[dict_learning])
cut_coords = plotting.find_xyz_cut_coords(cut_components)

for estimator, components in zip(estimators, components_img):
    plotting.plot_prob_atlas(components, view_type='filled_contours',
                            title='{0}'.format(names[estimator],
                                              cut_coords, colorbar=False))
    
    plotting.plot_stat_map(image.index_img(components, indices[estimator]),
                          title='{0}'.format(names[estimator],
                                            cut_coords=cut_coords, colorbar=False))
```

```python
dict_components = image.load_img('DictLearning_resting_stat.nii.gz')
canica_components = image.load_img('CanICA_resting_stat.nii.gz')
```

```python
plotting.plot_prob_atlas(dict_components)
plotting.plot_prob_atlas(canica_components)
```

```python
for img, cur_image in enumerate(image.iter_img(dict_components)):
    plotting.plot_stat_map(cur_img, title='IC{0} dict_learning'.format(img))
    plotting.plot_stat_map(image.index_img(canica_components, img), title='IC{0} ica_learning'.format(img))

    
```
